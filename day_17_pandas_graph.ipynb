#### Graph Function
1. plot                    - Simple Line Plot
2. plot.area               - Filled Version of Line Plot
3. plot.bar / plot.barh    - Bar Graph
4. plot.box                - Box Plot
5. plot.density            - Density Line (Smooth Curve)
6. plot.hexbin             - Hexagon Version of Scatter with Density
7. plot.hist               - Histogram
8. plot.kde                - Kernel Density Estimator (Density Graph)
9. plot.pie                - Pie Chart of Categorical Data
10. plot.scatter           - Scatter Graph


```python
df2.plot(); plt.title('Line Plot'); plt.show()
df2.plot.area(); plt.title('Area Plot'); plt.show()
df2.plot.bar(); plt.title('Bar Plot'); plt.show()
df2.plot.box(); plt.title('Box Plot'); plt.show()
df2.plot.barh(); plt.title('Bar Horizontal Plot'); plt.show()
df2.plot.density(); plt.title('Density Plot'); plt.show()# density is same as KDE
df2.plot.kde(); plt.title('KDE Plot'); plt.show()
df2.plot.hist(); plt.title('Histogram Plot'); plt.show()

df2.plot.scatter(x='a',y='b'); plt.title('Scatter Plot'); plt.show()
df2.plot.hexbin(x='a',y='b',C='c',gridsize=6); plt.title('Hexbin Plot'); plt.show()
```

```python
fig = plt.figure(figsize=(20,20))                
plt.subplot(4,3,1); df2.plot(); plt.title('Line Plot') 
plt.subplot(4,3,2); df2.plot.area(); plt.title('Area Plot')
plt.subplot(4,3,3); df2.plot.bar(); plt.title('Bar Plot')
plt.subplot(4,3,4); df2.plot.box(); plt.title('Box Plot')
plt.subplot(4,3,5); df2.plot.barh(); plt.title('Bar Horizontal Plot')
plt.subplot(4,3,6); df2.plot.density(); plt.title('Density Plot') # density is same as KDE
plt.subplot(4,3,7); df2.plot.kde(); plt.title('KDE Plot')
plt.subplot(4,3,8); df2.plot.hist(); plt.title('Histogram Plot')

plt.subplot(4,3,9); df2.plot.scatter(x='a',y='b'); plt.title('Scatter Plot')
plt.subplot(4,3,10); df2.plot.hexbin(x='a',y='b',C='c',gridsize=6);
plt.title('Hexbin Plot')
```

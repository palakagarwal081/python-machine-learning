#Types of Graph
1. plot -          line graph between two parameters
2. bar -           bar graph of single parameter (optional - categorical data) 
3. hist -          histogram of a bunch of data i.e. fequency of single parameter 
4. scatter -       dot garph between two parameters 
5. boxplot / box - statistical overview in form of box for a single Parameter
                   mean,outliers,quantiles (25%,75%),IQR
                   (optional - categorical data)
6. stem -          more like bar graph, Lines instead of bars 
7. violinplot -    Violin shaped statistical overview graph
                   spread of data (more like histogram)
                   single parameter (optional - categorical data)

```python
x = np.linspace(0,5,11)
y=x**3
```

```python
x
```

```python
plt.bar(x,y)
plt.title('Bar Graph')
plt.show()

plt.hist(x)
plt.title('Histogram') 
plt.show()

plt.scatter(x,y)
plt.title('Scatter')
plt.grid()
plt.xlabel('x--->')
plt.ylabel('y--->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()

plt.stem(x,y)
plt.title('Stem')
plt.show()

plt.violinplot(x)
plt.title('Violin plot')
plt.show()
```

```python
y2 =x**4
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
plt.plot(x,y,lw = 3,marker = 'o',markerfacecolor = 'yellow',markersize = 13) # lw = line width marker = show dots/triangles
plt.plot(x,y2,marker = 'v',markersize = 10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```

```python
fig = plt.figure()
plt.plot(x,y,lw = 3,marker = 'o',markerfacecolor = 'yellow',markersize = 13) # lw = line width marker = show dots/triangles
plt.plot(x,y2,marker = 'v',markersize = 10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```

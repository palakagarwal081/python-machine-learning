```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

#### Categories
1. Categorical Plots (hue can be used | comparison with categorical data)
    a. countplot (single parameter)
    b. barplot (one parameter is categorical and one is numerical)
    c. boxplot
    d. violinplot
    e. swarmplot
2. Distribution Plots
    a. distplot
    b. displot
    c. joinplot 
    d. pairplot 
    e. kdeplot
    f. rugplot
    g. scatterplot
3. Matrix Plot
    a.heatmap
    b. clustermap
    c. pairplot (same as above one)
4. Regression Plot
    a. lmplot
    b. regplot
5. Grid Plots
    a. PairGrid
    b. JoinGrid
    c. FacetGrid

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe(include = 'all')
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='total_bill',data=tips)
```

```python
sns.countplot(x='day',data=tips, hue='sex') # hue is like groupby
```

```python
sns.barplot(x='day',y= 'total_bill', data=tips) #  calculate mean by default means show average (mean) of thursday
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day',y= 'total_bill', data=tips,estimator = sum) #changing by default parameter
                                                                # estimator require function like sum /mean 
```

```python
sns.barplot(x='day',y= 'total_bill', data=tips, hue = 'time' ,estimator = np.mean)
```

```python
sns.boxplot(x='day',y= 'total_bill', data=tips, hue = 'time')
# here black dots are outliers
```

```python
sns.violinplot(x='day',y= 'total_bill', data=tips, hue = 'smoker')
# peaks like thing is data spread
```

```python
sns.swarmplot(x='day',y= 'total_bill', data=tips, hue = 'time');plt.title('Swarm')
# swarm action is which bird follow 
# shows data in dot form 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

#### Categories
1. Categorical Plots (hue can be used | comparison with categorical data)
    a. countplot (single parameter)
    b. barplot (one parameter is categorical and one is numerical | calculate mean by default)
    c. boxplot
    d. violinplot
    e. swarmplot
2. Distribution Plots
    a. distplot
    b. displot
    c. joinplot 
    d. pairplot 
    e. kdeplot
    f. rugplot
    g. scatterplot
3. Matrix Plot
    a.heatmap
    b. clustermap
    c. pairplot (same as above one)
4. Regression Plot
    a. Lmplot
    b. regplot
5. Grid Plots
    a. PairGrid
    b. JoinGrid
    c. FacetGrid

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe(include = 'all')
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='total_bill',data=tips)
```

```python
sns.countplot(x='day',data=tips, hue='sex') # hue is like groupby
```

```python
sns.barplot(x='day',y= 'total_bill', data=tips) #  calculate mean by default means show average (mean) of thursday
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day',y= 'total_bill', data=tips,estimator = sum) #changing by default parameter
                                                                # estimator require function like sum /mean 
```

```python
sns.barplot(x='day',y= 'total_bill', data=tips, hue = 'time' ,estimator = np.mean)
```

```python
sns.boxplot(x='day',y= 'total_bill', data=tips, hue = 'time')
# here black dots are outliers
```

```python
sns.violinplot(x='day',y= 'total_bill', data=tips, hue = 'smoker')
# peaks like thing is data spread
```

```python
sns.swarmplot(x='day',y= 'total_bill', data=tips, hue = 'time');plt.title('Swarm')
# swarm action is which bird follow 
# shows data in dot form 
```

```python
import warnings
warnings.filterwarnings('ignore')

sns.distplot(tips['total_bill'])
```

```python
sns.displot(data = tips, x='total_bill') # histogram
```

```python
sns.displot(data = tips, x='total_bill',y='tip') # kind of scatter plot
```

```python
sns.jointplot(data=tips,x='total_bill', y='tip')
plt.show()
sns.jointplot(data=tips,x='total_bill', y='tip',hue='smoker')
plt.show()
sns.jointplot(data=tips,x='total_bill', y='tip',kind ='hex') # kind hex show graph in hexagon form instead of dots(scatter)
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill' ,y='tip' )
```

```python
sns.kdeplot(data=tips,x='total_bill')
plt.show()
sns.kdeplot(data=tips,x='total_bill', y='tip')
plt.show()
```

```python
fig = plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
sns.kdeplot(data=tips,x='total_bill')
plt.subplot(1,2,2)
sns.kdeplot(data=tips,x='total_bill', y='tip')
```

```python
fig = plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
plt.subplot(1,2,2)
sns.kdeplot(data=tips,x='total_bill', y='tip')
sns.rugplot(data=tips,x='total_bill',y='tip')
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(data = iris)
```

```python
sns.pairplot(data = iris,hue='species')
```

```python
sns.pairplot(data = iris,hue='species', kind ='reg')  # kind='reg' means show regression line
```

```python
sns.regplot(data =tips,x='total_bill' , y='tip'); plt.title('regplot')  #show regression line
plt.show()
sns.lmplot(data = tips ,x = 'total_bill', y='tip'); plt.title('lmplot')  
```

```python
pg = sns.PairGrid(data = iris)
pg.map_diag(sns.kdeplot)
pg.map_lower(sns.scatterplot)
pg.map_upper(sns.kdeplot)
```

```python
jg = sns.JointGrid(data=tips , x='total_bill' , y='tip')
jg.plot(sns.regplot,sns.distplot)
```

```python
fg = sns.FacetGrid(data=tips, row = 'smoker', col = 'day',hue = 'sex')
fg.map(sns.scatterplot,'total_bill','tip').add_legend()
```
